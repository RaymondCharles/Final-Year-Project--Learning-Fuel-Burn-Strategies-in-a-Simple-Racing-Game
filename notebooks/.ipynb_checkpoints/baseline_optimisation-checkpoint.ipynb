# Baseline Optimisation Notebook

This notebook implements the base simulator, manual baseline, dynamic-programming benchmark, fixed-friction SLSQP optimiser, and drag-sensitive optimisation extension.

```python
# Defines the base RacingFuelEnv simulator used in the fixed-friction experiments.
# The environment tracks the current round, remaining fuel, and carried-over energy.
# The step() method applies the core dynamics: burn fuel, update energy, calculate distance,
# apply fixed friction, and return the next state.
import numpy as np

class RacingFuelEnv:
    """
    Simple fixed-horizon racing fuel environment.

    State:
      - t: current round (0..horizon)
      - fuel: fuel remaining
      - energy: kinetic-energy-like quantity (speed^2 is proportional to energy)

    Action:
      - burn: litres of fuel to burn this round (0..min(max_burn, fuel))

    Dynamics per round:
      1) Convert burn -> energy:  E_after = energy + burn * energy_per_litre
      2) Distance this round is proportional to speed: distance = sqrt(E_after)
      3) Apply friction between rounds: energy_next = friction * E_after
      4) fuel decreases: fuel_next = fuel - burn
    """
    def __init__(
        self,
        horizon=10,
        init_fuel=100.0,
        energy_per_litre=1.0,
        friction=0.90,         # 0<friction<1 means you lose energy each round
        max_burn=30.0,
        seed=None
    ):
        self.horizon = int(horizon)
        self.init_fuel = float(init_fuel)
        self.energy_per_litre = float(energy_per_litre)
        self.friction = float(friction)
        self.max_burn = float(max_burn)
        self.rng = np.random.default_rng(seed)
        self.reset()

    def reset(self):
        self.t = 0
        self.fuel = self.init_fuel
        self.energy = 0.0
        return self._obs()

    def _obs(self):
        # observation vector: [t, fuel, energy]
        return np.array([self.t, self.fuel, self.energy], dtype=np.float32)

    def step(self, burn):
        if self.t >= self.horizon:
            raise RuntimeError("Episode already finished. Call reset().")

        # Clip action to valid range
        burn = float(burn)
        burn = max(0.0, burn)
        burn = min(burn, self.max_burn)
        burn = min(burn, self.fuel)

        # 1) convert fuel to energy
        e_after = self.energy + burn * self.energy_per_litre

        # 2) reward is distance this round (speed), dt=1
        speed = np.sqrt(max(e_after, 0.0))
        reward = float(speed)

        # 3) friction loss between rounds
        self.energy = self.friction * e_after

        # 4) update fuel + time
        self.fuel -= burn
        self.t += 1

        done = (self.t >= self.horizon)
        info = {"burn": burn, "speed": speed, "energy_after": e_after}
        return self._obs(), reward, done, info
```

```python
# Runs the manual human-designed baseline policy through the base environment.
# This baseline provides an intuitive comparison point before using DP or optimisation.
# The cell logs burn, speed, fuel remaining, reward, and cumulative distance,
# then plots the manual strategy behaviour across the 10-round race.
import numpy as np
import matplotlib.pyplot as plt

env = RacingFuelEnv(horizon=10, init_fuel=100, friction=0.90, energy_per_litre=1.0, max_burn=30)
obs = env.reset()

# Manual baseline: front-loaded but gradually tapering fuel use
burn_plan = [20, 20, 20, 10, 10, 5, 5, 5, 3, 2]  # 10 actions for 10 rounds
H = env.horizon  # 10

history = {"t": [], "fuel": [], "energy": [], "speed": [], "burn": [], "reward": []}

# log initial state at t=0 (before any burning) 
history["t"].append(0)
history["fuel"].append(float(obs[1]))      # 100
history["energy"].append(float(obs[2]))    # 0
history["speed"].append(0.0)               # define start speed as 0
history["burn"].append(0.0)                # no burn yet
history["reward"].append(0.0)              # no distance yet

total = 0.0
done = False

# Run the 10 rounds, log each step as t=1..10 
for i, b in enumerate(burn_plan, start=1):
    obs, r, done, info = env.step(b)
    total += r

    history["t"].append(i)                 # 1..10
    history["fuel"].append(float(obs[1]))
    history["energy"].append(float(obs[2]))
    history["speed"].append(float(info["speed"]))
    history["burn"].append(float(info["burn"]))
    history["reward"].append(float(r))

    if done:
        break

print("Total distance (sum of rewards):", total)

# Plotting (now x-axis is truly 0..10) 
x = np.array(history["t"])  # 0..10

# Burn per round 
plt.figure()
plt.plot(x, history["burn"], marker="o")
plt.title("Burn per round")
plt.xlabel("round")
plt.ylabel("litres burned")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, env.max_burn)
plt.yticks(np.arange(0, int(env.max_burn) + 1, 5))
plt.grid(True, alpha=0.3)
plt.show()

# Speed per round 
plt.figure()
plt.plot(x, history["speed"], marker="o")
plt.title("Speed per round")
plt.xlabel("round")
plt.ylabel("speed")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
max_speed = max(history["speed"]) if history["speed"] else 1.0
plt.ylim(0, max_speed * 1.1)
plt.grid(True, alpha=0.3)
plt.show()

# Fuel remaining 
plt.figure()
plt.plot(x, history["fuel"], marker="o")
plt.title("Fuel remaining")
plt.xlabel("round")
plt.ylabel("litres")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, env.init_fuel)
plt.yticks(np.arange(0, int(env.init_fuel) + 1, 10))
plt.grid(True, alpha=0.3)
plt.show()

# Cumulative distance 
plt.figure()
cumdist = np.cumsum(history["reward"])
plt.plot(x, cumdist, marker="o")
plt.title("Cumulative distance")
plt.xlabel("round")
plt.ylabel("distance (cumulative)")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, float(cumdist[-1]) * 1.05)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Defines the dynamic-programming solver for the fixed-friction environment.
# Because energy is continuous, energy is discretised into bins so that a finite DP table can be built.
# The algorithm uses backward induction to find the best burn action for each round, fuel level,
# and discretised energy state.
def dp_optimal_policy(
    horizon=10,
    init_fuel=100,
    friction=0.90,
    energy_per_litre=1.0,
    max_burn=30,
    e_bin=1.0,
    e_max=1000.0
):
    """
    Discretised DP to compute an (approx) optimal policy.

    Returns:
      V: value table
      pi: greedy policy table (burn choice) indexed by (t, fuel, energy_bin)
    """
    H = horizon
    F = int(init_fuel)
    max_burn = int(max_burn)
    e_bin = float(e_bin)

    nE = int(e_max / e_bin) + 1

    # V[t, fuel, e] = best achievable remaining distance from time t onward
    V = np.zeros((H + 1, F + 1, nE), dtype=np.float32)
    pi = np.zeros((H, F + 1, nE), dtype=np.int32)

    def e_to_idx(e):
        e = max(0.0, min(e, e_max))
        return int(round(e / e_bin))

    # Work backwards from the final round to the first round.
    # For each possible state, test all feasible burn actions and keep the one
    # that gives the highest immediate reward plus future value.
    for t in range(H - 1, -1, -1):
        for fuel in range(F + 1):
            for e_i in range(nE):
                e = e_i * e_bin

                best = -1e9
                best_a = 0

                # Try all feasible burns
                a_max = min(max_burn, fuel)
                for burn in range(a_max + 1):
                    e_after = e + burn * energy_per_litre
                    reward = np.sqrt(e_after)
                    e_next = friction * e_after
                    e_next_i = e_to_idx(e_next)

                    val = reward + V[t + 1, fuel - burn, e_next_i]
                    if val > best:
                        best = val
                        best_a = burn

                V[t, fuel, e_i] = best
                pi[t, fuel, e_i] = best_a

    return V, pi

def simulate_with_policy(pi, env):
    obs = env.reset()
    total = 0.0

    hist = {"burn": [], "speed": [], "fuel": [], "energy": [], "reward": []}

    done = False
    while not done:
        t = int(obs[0])
        fuel = int(round(obs[1]))
        energy = float(obs[2])

        # Match DP's energy binning assumptions (same as dp_optimal_policy call)
        # We'll store these when we call dp to avoid mismatch.
        raise RuntimeError("Call simulate_with_policy_configured instead.")
```

```python
# Simulates a policy produced by the DP solver.
# The policy table pi maps each discretised state to the recommended fuel burn.
# This cell replays the DP policy in the original environment so the approximate DP result
# can be checked through a forward simulation.
def simulate_with_policy_configured(pi, env, e_bin=1.0, e_max=1000.0):
    def e_to_idx(e):
        e = max(0.0, min(e, e_max))
        return int(round(e / e_bin))

    obs = env.reset()
    total = 0.0

    hist = {"t": [], "burn": [], "speed": [], "fuel": [], "energy": [], "reward": []}

    # initial point at t=0
    hist["t"].append(0)
    hist["burn"].append(0.0)
    hist["speed"].append(0.0)
    hist["fuel"].append(float(obs[1]))
    hist["energy"].append(float(obs[2]))
    hist["reward"].append(0.0)

    done = False
    step_count = 0

    while not done:
        t = int(obs[0])
        fuel = int(round(obs[1]))
        energy = float(obs[2])
        e_i = e_to_idx(energy)

        burn = int(pi[t, fuel, e_i])

        obs, r, done, info = env.step(burn)
        total += r
        step_count += 1

        hist["t"].append(step_count)
        hist["burn"].append(float(info["burn"]))
        hist["speed"].append(float(info["speed"]))
        hist["fuel"].append(float(obs[1]))
        hist["energy"].append(float(obs[2]))
        hist["reward"].append(float(r))

    return total, hist
```

```python
# Runs the fixed-friction dynamic-programming benchmark and plots the resulting policy.
# This cell computes the DP value table and policy, simulates the recommended policy,
# prints the approximate optimal return, and produces plots for burn, speed, fuel, and distance.
import numpy as np
import matplotlib.pyplot as plt

# -Configuration
H = 10
F = 100
friction = 0.90
epl = 1.0
max_burn = 30
e_bin = 1.0
e_max = 1000.0

# Checks
missing = []
for name in ["RacingFuelEnv", "dp_optimal_policy", "simulate_with_policy_configured"]:
    if name not in globals():
        missing.append(name)
if missing:
    raise NameError(f"Missing definitions: {missing}. Run the cells that define them first.")

# Compute the DP value table V and policy table pi using the configured discretisation.
V, pi = dp_optimal_policy(
    horizon=H,
    init_fuel=F,
    friction=friction,
    energy_per_litre=epl,
    max_burn=max_burn,
    e_bin=e_bin,
    e_max=e_max
)

# Simulate the DP Policy
env = RacingFuelEnv(
    horizon=H,
    init_fuel=F,
    friction=friction,
    energy_per_litre=epl,
    max_burn=max_burn
)

# Replay the DP policy in the simulator to verify the policy through forward simulation.
total_opt, hist_opt = simulate_with_policy_configured(pi, env, e_bin=e_bin, e_max=e_max)

print("Keys in hist_opt:", hist_opt.keys())   # should include 't'
print("DP approx-optimal total distance:", total_opt)

# Plots (4 graphs, full axes) 
x = np.array(hist_opt["t"])  # should be 0..10

# Burn per round
plt.figure()
plt.plot(x, hist_opt["burn"], marker="o")
plt.title("DP policy: burn per round")
plt.xlabel("round")
plt.ylabel("litres burned")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, max_burn)
plt.yticks(np.arange(0, max_burn + 1, 5))
plt.grid(True, alpha=0.3)
plt.show()

# Speed per round
plt.figure()
plt.plot(x, hist_opt["speed"], marker="o")
plt.title("DP policy: speed per round")
plt.xlabel("round")
plt.ylabel("speed")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
max_speed = max(hist_opt["speed"]) if hist_opt["speed"] else 1.0
plt.ylim(0, max_speed * 1.1)
plt.grid(True, alpha=0.3)
plt.show()

# Fuel remaining
plt.figure()
plt.plot(x, hist_opt["fuel"], marker="o")
plt.title("DP policy: fuel remaining")
plt.xlabel("round")
plt.ylabel("litres")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, F)
plt.yticks(np.arange(0, F + 1, 10))
plt.grid(True, alpha=0.3)
plt.show()

# Cumulative distance
plt.figure()
cumdist = np.cumsum(hist_opt["reward"])
plt.plot(x, cumdist, marker="o")
plt.title("DP policy: cumulative distance")
plt.xlabel("round")
plt.ylabel("distance (cumulative)")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, float(cumdist[-1]) * 1.05)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Compares several fixed-friction policies against each other.
# The policies include the manual baseline, even fuel use, conservative fuel use,
# front-loaded fuel use, and the DP policy. This helps show why the DP policy is better
# than simple intuitive alternatives.
import numpy as np
import matplotlib.pyplot as plt


# Helper: replay a fixed burn schedule and collect the trajectory for plotting.
def simulate_with_burn_plan(env, burn_plan):
    obs = env.reset()
    total = 0.0

    hist = {
        "t": [0],
        "burn": [0.0],
        "speed": [0.0],
        "fuel": [float(obs[1])],     # should start at init fuel (100)
        "energy": [float(obs[2])],   # should start at 0
        "reward": [0.0]
    }

    done = False
    step_count = 0

    for b in burn_plan:
        if done:
            break

        obs, r, done, info = env.step(b)
        total += float(r)
        step_count += 1

        hist["t"].append(step_count)              # 1..H
        hist["burn"].append(float(info["burn"]))
        hist["speed"].append(float(info["speed"]))
        hist["fuel"].append(float(obs[1]))
        hist["energy"].append(float(obs[2]))
        hist["reward"].append(float(r))

    return total, hist

# Helper: replay a fixed burn schedule and collect the trajectory for plotting.
def pad_to_len(arr, length):
    arr = list(arr)
    if len(arr) < length:
        arr += [arr[-1] if len(arr) else 0.0] * (length - len(arr))
    return np.array(arr[:length], dtype=float)


# Checks
missing = []
for name in ["RacingFuelEnv", "dp_optimal_policy", "simulate_with_policy_configured"]:
    if name not in globals():
        missing.append(name)

if missing:
    raise NameError(f"Missing definitions: {missing}. Run earlier cells first (env + DP + simulate_with_policy_configured).")


# Environment Configuration
H = 10
F = 100
friction = 0.90
epl = 1.0
max_burn = 30
e_bin = 1.0
e_max = 1000.0


# Define simple comparison policies to judge the DP result against intuitive baselines.
manual_plan = [20, 20, 20, 10, 10, 5, 5, 5, 3, 2]
even_plan = [F / H] * H
conservative_plan = [5] * H
frontload_plan = [max_burn] * H

policies = [
    ("Conservative (5 each)", conservative_plan),
    ("Even burn (F/H)", even_plan),
    ("Front-load (max each)", frontload_plan),
    ("My plan", manual_plan),
]


# Run baselines 
results = []

for name, plan in policies:
    env = RacingFuelEnv(horizon=H, init_fuel=F, friction=friction, energy_per_litre=epl, max_burn=max_burn)
    total, hist = simulate_with_burn_plan(env, plan)
    results.append({
        "name": name,
        "total_distance": total,
        "final_fuel": hist["fuel"][-1] if hist["fuel"] else F,
        "peak_speed": float(np.max(hist["speed"])) if hist["speed"] else 0.0,
        "hist": hist
    })


# Run DP benchmark 
V, pi = dp_optimal_policy(
    horizon=H,
    init_fuel=F,
    friction=friction,
    energy_per_litre=epl,
    max_burn=max_burn,
    e_bin=e_bin,
    e_max=e_max
)

env = RacingFuelEnv(horizon=H, init_fuel=F, friction=friction, energy_per_litre=epl, max_burn=max_burn)
dp_total, dp_hist = simulate_with_policy_configured(pi, env, e_bin=e_bin, e_max=e_max)


if "t" not in dp_hist:
    dp_hist["t"] = list(range(len(dp_hist["burn"])))

if len(dp_hist["t"]) == H and (len(dp_hist["burn"]) == H):
    dp_hist["t"] = [0] + [i for i in range(1, H + 1)]
    dp_hist["burn"] = [0.0] + list(dp_hist["burn"])
    dp_hist["speed"] = [0.0] + list(dp_hist["speed"])

    dp_hist["fuel"] = [float(F)] + list(dp_hist["fuel"])
    dp_hist["energy"] = [0.0] + list(dp_hist["energy"])
    dp_hist["reward"] = [0.0] + list(dp_hist["reward"])

results.append({
    "name": "DP approx-optimal",
    "total_distance": dp_total,
    "final_fuel": dp_hist["fuel"][-1] if dp_hist["fuel"] else F,
    "peak_speed": float(np.max(dp_hist["speed"])) if dp_hist["speed"] else 0.0,
    "hist": dp_hist
})


# Print comparison table 
results_sorted = sorted(results, key=lambda d: d["total_distance"], reverse=True)
best = results_sorted[0]["total_distance"]

print("\n=== Policy comparison (higher total distance is better) ===")
print(f"Environment: H={H}, Fuel={F}, friction={friction}, energy_per_litre={epl}, max_burn={max_burn}\n")

header = f"{'Policy':30s} {'TotalDist':>10s} {'GapToBest':>10s} {'FinalFuel':>10s} {'PeakSpeed':>10s}"
print(header)
print("-" * len(header))

for r in results_sorted:
    gap = best - r["total_distance"]
    print(f"{r['name'][:30]:30s} {r['total_distance']:10.3f} {gap:10.3f} {r['final_fuel']:10.3f} {r['peak_speed']:10.3f}")


# Plots (4 graphs) 
x = np.arange(H + 1)  # 0..10

# Burn schedule comparison
plt.figure()
for r in results_sorted:
    burn = pad_to_len(r["hist"]["burn"], H + 1)
    plt.plot(x, burn, label=r["name"])
plt.title("Burn schedule comparison")
plt.xlabel("round")
plt.ylabel("litres burned")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, max_burn)
plt.yticks(np.arange(0, int(max_burn) + 1, 5))
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# Speed per round comparison
plt.figure()
max_speed_overall = 0.0
for r in results_sorted:
    speed = pad_to_len(r["hist"]["speed"], H + 1)
    max_speed_overall = max(max_speed_overall, float(np.max(speed)))
    plt.plot(x, speed, label=r["name"])
plt.title("Speed per round comparison")
plt.xlabel("round")
plt.ylabel("speed")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, max_speed_overall * 1.1 if max_speed_overall > 0 else 1.0)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# Fuel remaining comparison
plt.figure()
for r in results_sorted:
    fuel = pad_to_len(r["hist"]["fuel"], H + 1)
    plt.plot(x, fuel, label=r["name"])
plt.title("Fuel remaining comparison")
plt.xlabel("round")
plt.ylabel("litres")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, F)
plt.yticks(np.arange(0, int(F) + 1, 10))
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# Cumulative distance comparison
plt.figure()
for r in results_sorted:
    rewards = pad_to_len(r["hist"]["reward"], H + 1)
    cumdist = np.cumsum(rewards)
    plt.plot(x, cumdist, label=r["name"])
plt.title("Cumulative distance comparison")
plt.xlabel("round")
plt.ylabel("distance (cumulative)")
plt.xlim(0, H)
plt.xticks(np.arange(0, H + 1, 1))
plt.ylim(0, float(np.max(cumdist)) * 1.05)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
# Runs a faster, lower-resolution DP sensitivity experiment.
# This cell is not the main final benchmark; it is used to compare behaviour quickly
# under different fixed-friction values and to check that the policy structure is sensible
import numpy as np
import time

# Uses coarser energy bins and larger burn steps to reduce runtime.
# This trades accuracy for speed and is mainly used for sensitivity checks.
def dp_optimal_policy_fast(
    horizon=10,
    init_fuel=100,
    friction=0.90,
    energy_per_litre=1.0,
    max_burn=30,
    burn_step=5,    
    e_bin=10.0,       
    e_max=800.0       
):

    H = int(horizon)
    F = int(init_fuel)
    max_burn = int(max_burn)
    burn_step = int(burn_step)
    e_bin = float(e_bin)
    e_max = float(e_max)

    # Energy bins
    nE = int(e_max / e_bin) + 1

    V = np.zeros((H + 1, F + 1, nE), dtype=np.float32)
    pi = np.zeros((H, F + 1, nE), dtype=np.int32)

    # Precompute burn candidates
    burn_candidates = np.arange(0, max_burn + 1, burn_step, dtype=np.int32)

    def e_to_idx(e):
        e = np.clip(e, 0.0, e_max)
        return np.rint(e / e_bin).astype(np.int32)

    # Backward induction
    for t in range(H - 1, -1, -1):
        for fuel in range(F + 1):
            feasible = burn_candidates[burn_candidates <= fuel]  # Burns allowed by fuel
            if feasible.size == 0:
                feasible = np.array([0], dtype=np.int32)

            # Vector stuff independent of energy state:
            fuel_next = (fuel - feasible).astype(np.int32)  # Index into V

            for e_i in range(nE):
                e = e_i * e_bin

                # Vectorised over feasible burns
                e_after = e + feasible * energy_per_litre
                reward = np.sqrt(np.maximum(e_after, 0.0))
                e_next = friction * e_after
                e_next_i = e_to_idx(e_next)

                # Gather V[t+1, fuel-burn, e_next_i] for each burn
                future = V[t + 1, fuel_next, e_next_i]
                vals = reward + future

                best_idx = int(np.argmax(vals))
                V[t, fuel, e_i] = float(vals[best_idx])
                pi[t, fuel, e_i] = int(feasible[best_idx])

    return V, pi


# Fast sensitivity test 
H = 10
F = 100
epl = 1.0
max_burn = 30

# Test how different fixed-friction levels affect the best fuel-use pattern.
frictions = [0.80, 0.90, 0.95]

manual_plan = [20, 20, 20, 10, 10, 5, 5, 5, 3, 2]
even_plan = [F / H] * H
conservative_plan = [5] * H
frontload_plan = [max_burn] * H

policies = [
    ("Conservative (5 each)", conservative_plan),
    ("Even burn (F/H)", even_plan),
    ("Front-load (max each)", frontload_plan),
    ("My plan", manual_plan),
]

# Checks
missing = []
for name in ["RacingFuelEnv", "simulate_with_policy_configured"]:
    if name not in globals():
        missing.append(name)
if missing:
    raise NameError(f"Missing definitions: {missing}. Run env + simulate_with_policy_configured first.")

def simulate_total(env, burn_plan):
    obs = env.reset()
    total = 0.0
    done = False
    for b in burn_plan:
        if done:
            break
        obs, r, done, info = env.step(b)
        total += float(r)
    return total

print("=== FAST sensitivity study (vectorised DP + coarser actions) ===")
print(f"H={H}, Fuel={F}, max_burn={max_burn}")
print("DP settings: burn_step=5, e_bin=10, e_max=800\n")

for idx, fr in enumerate(frictions, start=1):
    t0 = time.perf_counter()

    totals = {}
    for name, plan in policies:
        env = RacingFuelEnv(horizon=H, init_fuel=F, friction=fr, energy_per_litre=epl, max_burn=max_burn)
        totals[name] = simulate_total(env, plan)

    # Fast DP
    V, pi = dp_optimal_policy_fast(
        horizon=H, init_fuel=F, friction=fr, energy_per_litre=epl,
        max_burn=max_burn, burn_step=5, e_bin=10.0, e_max=800.0
    )

    env = RacingFuelEnv(horizon=H, init_fuel=F, friction=fr, energy_per_litre=epl, max_burn=max_burn)
    dp_total, _ = simulate_with_policy_configured(pi, env, e_bin=10.0, e_max=800.0)
    totals["DP approx-optimal (fast)"] = dp_total

    best_name = max(totals, key=totals.get)
    best_val = totals[best_name]

    print(f"--- ({idx}/{len(frictions)}) friction = {fr:.2f} ---")
    print(f"Best: {best_name}  TotalDist={best_val:.3f}")
    for name, val in sorted(totals.items(), key=lambda x: x[1], reverse=True):
        gap = best_val - val
        pct = 100.0 * val / best_val
        print(f"  {name:26s}  {val:8.3f}  gap={gap:7.3f}  {pct:6.1f}%")
    print(f"Time for this friction: {time.perf_counter()-t0:.2f}s\n")
```

```python
# Computes the strongest fixed-friction baseline using constrained optimisation.
# The burn schedule is treated as the optimisation variable, and SLSQP searches
# for the schedule that maximises total distance while using exactly 100 litres
# and respecting the 30-litre per-round cap.
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Configuration
HORIZON = 10
TOTAL_FUEL = 100.0
FRICTION = 0.90
MAX_BURN_PER_ROUND = 30.0
ENERGY_PER_LITRE = 1.0  

# Validation
if TOTAL_FUEL < 0:
    raise ValueError("TOTAL_FUEL must be nonnegative.")
if not (0.0 < FRICTION <= 1.0):
    raise ValueError("FRICTION must be in (0, 1].")
if MAX_BURN_PER_ROUND <= 0:
    raise ValueError("MAX_BURN_PER_ROUND must be positive.")
if TOTAL_FUEL > HORIZON * MAX_BURN_PER_ROUND + 1e-9:
    raise ValueError(
        "Infeasible: TOTAL_FUEL is larger than HORIZON * MAX_BURN_PER_ROUND, "
        "so you cannot spend all fuel within the per-round cap."
    )

# Forward-simulate a full burn schedule under the fixed-friction dynamics.
def rollout(burn_schedule):
    """
    Returns a dict with time-series that align to t = 0..HORIZON:
      - burn[t] is defined with burn[0]=0 (no action at t=0),
        and burn[1..H] correspond to rounds 1..H.
      - speed[t] is defined similarly with speed[0]=0.
      - energy[t] is the carried energy *after* friction at the end of round t
        (energy[0]=0 initial).
      - fuel[t] is remaining fuel after spending through round t (fuel[0]=TOTAL_FUEL).
      - reward[t] is distance per round (reward[0]=0).
    """
    burns = np.asarray(burn_schedule, dtype=float)
    if burns.shape != (HORIZON,):
        raise ValueError(f"burn_schedule must have shape ({HORIZON},).")

    # Time axis 0..H
    t = np.arange(0, HORIZON + 1)

    burn_ts = np.zeros(HORIZON + 1, dtype=float)
    speed_ts = np.zeros(HORIZON + 1, dtype=float)
    reward_ts = np.zeros(HORIZON + 1, dtype=float)
    energy_ts = np.zeros(HORIZON + 1, dtype=float)
    fuel_ts = np.zeros(HORIZON + 1, dtype=float)

    fuel_ts[0] = TOTAL_FUEL
    energy = 0.0

    for i in range(1, HORIZON + 1):
        b = float(burns[i - 1])

        # Safety clip (bounds + equality constraint should already ensure validity)
        b = np.clip(b, 0.0, MAX_BURN_PER_ROUND)

        e_after = energy + b * ENERGY_PER_LITRE
        speed = np.sqrt(max(e_after, 0.0))
        reward = float(speed)

        # Apply friction between rounds
        energy = FRICTION * e_after

        burn_ts[i] = b
        speed_ts[i] = speed
        reward_ts[i] = reward
        energy_ts[i] = energy
        fuel_ts[i] = TOTAL_FUEL - np.sum(burns[:i])  # remaining after i rounds

    total_distance = float(np.sum(reward_ts))  # reward_ts[0]=0 so this is sum over rounds 1..H
    return {
        "t": t,
        "burn": burn_ts,
        "speed": speed_ts,
        "reward": reward_ts,
        "energy": energy_ts,
        "fuel": fuel_ts,
        "total_distance": total_distance,
    }

# Objective for SLSQP:
# return negative total distance because scipy.optimize.minimize performs minimisation.
def objective_negative_total_distance(burn_schedule):
    """
    SciPy minimizes. We want to maximize total distance, so return negative distance.
    """
    energy = 0.0
    total_distance = 0.0

    for b in burn_schedule:
        e_after = energy + float(b) * ENERGY_PER_LITRE
        total_distance += np.sqrt(max(e_after, 0.0))
        energy = FRICTION * e_after

    return -float(total_distance)

# Optimization Setup
# Initial guess: uniform spending (feasible because TOTAL_FUEL <= HORIZON * MAX_BURN_PER_ROUND)
x0 = np.full(HORIZON, TOTAL_FUEL / HORIZON, dtype=float)

# Bounds: 0 <= burn_t <= MAX_BURN_PER_ROUND
bounds = [(0.0, MAX_BURN_PER_ROUND) for _ in range(HORIZON)]

# Equality constraint: sum(burns) == TOTAL_FUEL
constraints = [
    {"type": "eq", "fun": lambda x: np.sum(x) - TOTAL_FUEL}
]

# Run SLSQP with equality constraint on total fuel and bounds on each round's burn.-
result = minimize(
    fun=objective_negative_total_distance,
    x0=x0,
    method="SLSQP",
    bounds=bounds,
    constraints=constraints,
    options={"ftol": 1e-12, "maxiter": 2000, "disp": True},
)

if not result.success:
    raise RuntimeError(
        "Optimization failed.\n"
        f"status={result.status}\n"
        f"message={result.message}\n"
        f"x={result.x}\n"
    )

optimal_burns = result.x
max_dist = -result.fun

print("\n=== SLSQP / Lagrange-baseline result ===")
print(f"Optimal total distance: {max_dist:.6f}")
print("Optimal burn schedule (rounds 1..H):", np.round(optimal_burns, 6))
print(f"Sum burns: {optimal_burns.sum():.12f} (target {TOTAL_FUEL})")

# SLSQP reports constraint multipliers in OptimizeResult.multipliers (when available)
if hasattr(result, "multipliers"):
    print("Constraint multipliers (KKT/Lagrange-style):", result.multipliers)

# Plotting & Verification
traj = rollout(optimal_burns)

t = traj["t"]  # 0..H
burn_ts = traj["burn"]
speed_ts = traj["speed"]
fuel_ts = traj["fuel"]
cumdist_ts = np.cumsum(traj["reward"])

plt.figure(figsize=(12, 8))

# Burn per round
ax1 = plt.subplot(2, 2, 1)
ax1.plot(t, burn_ts, marker="o")
ax1.set_title("Burn per round (SLSQP baseline)")
ax1.set_xlabel("Round")
ax1.set_ylabel("Litres burned")
ax1.set_xlim(0, HORIZON)
ax1.set_ylim(0, MAX_BURN_PER_ROUND * 1.05)
ax1.set_xticks(np.arange(0, HORIZON + 1, 1))
ax1.grid(True, alpha=0.3)

# Speed per round
ax2 = plt.subplot(2, 2, 2)
ax2.plot(t, speed_ts, marker="o")
ax2.set_title("Speed per round")
ax2.set_xlabel("Round")
ax2.set_ylabel("Speed (distance/round)")
ax2.set_xlim(0, HORIZON)
ax2.set_xticks(np.arange(0, HORIZON + 1, 1))
ax2.set_ylim(0, max(speed_ts) * 1.10 if np.max(speed_ts) > 0 else 1.0)
ax2.grid(True, alpha=0.3)

# Fuel remaining
ax3 = plt.subplot(2, 2, 3)
ax3.plot(t, fuel_ts, marker="o")
ax3.set_title("Fuel remaining")
ax3.set_xlabel("Round")
ax3.set_ylabel("Litres")
ax3.set_xlim(0, HORIZON)
ax3.set_xticks(np.arange(0, HORIZON + 1, 1))
ax3.set_ylim(0, TOTAL_FUEL * 1.05)
ax3.grid(True, alpha=0.3)

# Cumulative distance
ax4 = plt.subplot(2, 2, 4)
ax4.plot(t, cumdist_ts, marker="o")
ax4.set_title("Cumulative distance")
ax4.set_xlabel("Round")
ax4.set_ylabel("Distance (cumulative)")
ax4.set_xlim(0, HORIZON)
ax4.set_xticks(np.arange(0, HORIZON + 1, 1))
ax4.set_ylim(0, cumdist_ts[-1] * 1.05 if cumdist_ts[-1] > 0 else 1.0)
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Drag-sensitive extension: speed-dependent carry-over - Defines the drag-sensitive version of the racing environment.
# Unlike the fixed-friction model, the carry-over coefficient decreases as post-burn energy increases.
# This makes very aggressive early burning less efficient and creates a more meaningful pacing problem.
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Drag-sensitive model settings

HORIZON = 10
INIT_FUEL = 100
MAX_BURN = 30
ALPHA = 1.0

BASE_RETENTION = 0.95
DRAG_COEFF = 0.0035    
MIN_RETENTION = 0.55
MAX_RETENTION = 0.95

# Compute the energy carry-over coefficient.
# Higher post-burn energy gives lower retention, clipped between minimum and maximum limits.
def drag_retention(e_after,
                   base_retention=BASE_RETENTION,
                   drag_coeff=DRAG_COEFF,
                   min_retention=MIN_RETENTION,
                   max_retention=MAX_RETENTION):
    
    # Carry-over decreases as post-burn energy increases.
    
    rho = base_retention - drag_coeff * e_after
    return float(np.clip(rho, min_retention, max_retention))


# Simulate a full burn schedule under the drag-sensitive dynamics.
# This returns both total distance and a detailed per-round history.
def simulate_drag_plan(burns,
                       horizon=HORIZON,
                       init_fuel=INIT_FUEL,
                       max_burn=MAX_BURN,
                       alpha=ALPHA,
                       base_retention=BASE_RETENTION,
                       drag_coeff=DRAG_COEFF,
                       min_retention=MIN_RETENTION,
                       max_retention=MAX_RETENTION):
    
    fuel = float(init_fuel)
    energy = 0.0
    total_distance = 0.0
    history = []

    for t, burn in enumerate(burns):
        burn = float(np.clip(burn, 0.0, min(max_burn, fuel)))

        energy_after = energy + alpha * burn
        distance = math.sqrt(max(energy_after, 0.0))
        rho = drag_retention(
            energy_after,
            base_retention=base_retention,
            drag_coeff=drag_coeff,
            min_retention=min_retention,
            max_retention=max_retention
        )
        energy_next = rho * energy_after
        fuel_next = fuel - burn

        total_distance += distance

        history.append({
            "t": t + 1,
            "burn": burn,
            "fuel_before": fuel,
            "fuel_after": fuel_next,
            "energy_before": energy,
            "energy_after": energy_after,
            "distance": distance,
            "retention": rho,
            "energy_next": energy_next,
            "cumulative_distance": total_distance
        })

        fuel = fuel_next
        energy = energy_next

    return total_distance, history

# Objective for drag-sensitive SLSQP optimisation.
# The optimiser minimises this value, so it returns negative total distance.
def negative_total_distance_drag(burns):
    total_distance, _ = simulate_drag_plan(burns)
    return -total_distance

# Convert a continuous burn schedule into a whole-litre schedule while preserving the 100-litre budget.
# This is needed because the DQN can only choose one-litre actions.
def round_schedule_to_budget(x, total_budget=INIT_FUEL, cap=MAX_BURN):

    x = np.asarray(x, dtype=float)

    floor_vals = np.floor(x).astype(int)
    floor_vals = np.clip(floor_vals, 0, cap)

    shortfall = int(total_budget - floor_vals.sum())
    fractions = x - floor_vals
    order = np.argsort(-fractions)

    out = floor_vals.copy()
    k = 0
    while shortfall > 0:
        idx = order[k % len(order)]
        if out[idx] < cap:
            out[idx] += 1
            shortfall -= 1
        k += 1

    return out
```

```python
# Runs SLSQP on the drag-sensitive model.
# The continuous optimiser can choose fractional litre values.
# The resulting continuous schedule is then converted into a rounded integer-feasible benchmark
# so it can be compared fairly with the DQN, which uses one-litre actions.
# Initial guess: spread fuel evenly
x0 = np.full(HORIZON, INIT_FUEL / HORIZON, dtype=float)

# Bounds and budget constraint
bounds = [(0.0, MAX_BURN)] * HORIZON
constraints = [
    {"type": "eq", "fun": lambda x: np.sum(x) - INIT_FUEL}
]

result_drag = minimize(
    negative_total_distance_drag,
    x0=x0,
    method="SLSQP",
    bounds=bounds,
    constraints=constraints,
    options={"maxiter": 2000, "ftol": 1e-12, "disp": True}
)

drag_opt_burns_cont = np.clip(result_drag.x, 0.0, MAX_BURN)
# Convert the continuous fractional solution into a whole-litre benchmark for DQN comparison.
drag_opt_burns_int = round_schedule_to_budget(drag_opt_burns_cont)

drag_opt_dist_cont, drag_opt_hist_cont = simulate_drag_plan(drag_opt_burns_cont)
drag_opt_dist_int, drag_opt_hist_int = simulate_drag_plan(drag_opt_burns_int)

greedy_burns = np.array([30, 30, 30, 10, 0, 0, 0, 0, 0, 0], dtype=float)
uniform_burns = np.array([10] * 10, dtype=float)

greedy_dist, greedy_hist = simulate_drag_plan(greedy_burns)
uniform_dist, uniform_hist = simulate_drag_plan(uniform_burns)

print("=== Drag-sensitive SLSQP result ===")
print("Success:", result_drag.success)
print("Message:", result_drag.message)
print("Continuous optimum burns:", np.round(drag_opt_burns_cont, 3))
print("Integer-feasible benchmark:", drag_opt_burns_int.tolist())
print(f"Continuous optimum total distance: {drag_opt_dist_cont:.6f}")
print(f"Integer benchmark total distance: {drag_opt_dist_int:.6f}")
print(f"Greedy total distance: {greedy_dist:.6f}")
print(f"Uniform total distance: {uniform_dist:.6f}")
```

```python
# Clarify continuous optimum vs integer-feasible benchmark
# The continuous SLSQP optimum can use fractional litre values.
# The DQN later uses one-litre discrete actions, so we also create
# a rounded integer-feasible benchmark for fair DQN comparison.

print("=== Continuous optimum vs integer-feasible benchmark ===")
print("Continuous SLSQP optimum burns:")
print(np.round(drag_opt_burns_cont, 3))
print(f"Continuous optimum total distance: {drag_opt_dist_cont:.6f}")

print("\nRounded integer-feasible benchmark burns:")
print(drag_opt_burns_int.tolist())
print(f"Integer-feasible benchmark total distance: {drag_opt_dist_int:.6f}")

print("\nExplanation:")
print("The integer schedule is not the continuous optimum itself.")
print("It is a whole-litre benchmark used because the DQN action space is discrete.")
```
